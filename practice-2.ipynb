```python
import os

def quicksort(userlist):
    if len(userlist)<=1:
        return userlist
    pivot = userlist[len(userlist)//2]
    left = [x  for x in userlist if x < pivot]
    center = [x for x in userlist if x ==pivot]
    right = [x for x in userlist if x >pivot]
    return quicksort(left)+ center + quicksort(right)

testlist = [ 234, 13 , 454, 575 , 23 , 454]

print(quicksort(testlist))
```

```python
import numpy as np

def bubblesort(userlist):
    for i in range(len(userlist)):
        for j in range(0, len(userlist)-i-1):
            if userlist[j] > userlist[j+1]:
                temp = userlist[j]
                userlist[j] = userlist[j+1]
                userlist[j+1]=temp

    return userlist
testlist = [ 234, 13 , 454, 575 , 23 , 454]

print(bubblesort(testlist))

```

```python
import threading

counter:int =0
counter_lock = threading.Lock()

def incrementCounter(numberList):
    global counter
    for i in numberList:
        with counter_lock:
            counter += i

list1=[1,2,3,4,5,6]
list2=[4,5,6,7,8,9]

thread1 = threading.Thread(target = incrementCounter, args=(list1,))
thread2 = threading.Thread(target = incrementCounter, args=(list2,))

thread1.start()
thread2.start()

thread1.join()
thread2.join()

print("Final value of counter", counter)
```

```python
# merge sort algorithm
class mergesort():
    def __init__(self):
        pass

    def sort(self, _list):
        if len(_list)<=1:
           # print("Empty or single element list")
            return _list
        else:
            mid = len(_list)//2
            left = _list[:mid]
            right = _list[mid:]
            self.sort(left)
            self.sort(right)
            self.merge(left, right, _list)
            return _list
    
    def merge(self, left, right, _list):
        i=0
        j=0
        k=0
        while i<len(left) and j<len(right):
            if left[i]<right[j]:
                _list[k]=left[i]
                i+=1
            else:
                _list[k]=right[j]
                j+=1
            k+=1
        while i<len(left):
            _list[k]=left[i]
            i+=1
            k+=1
        while j<len(right):                    
            _list[k]=right[j]
            j+=1
            k+=1
        return _list
    
    def printlist(self, _list):
        for i in _list:
            print(i)

def main():
    list1 = [ 23, 45, 12, 34, 56, 78, 90, 23, 45, 67, 89]
    ms = mergesort()
    ms.sort(list1)
    ms.printlist(list1)

if __name__ == "__main__":
    main()
```

```python
# enumerate actions
import random
import matplotlib.pyplot as plt

#random.seed(124)

actions = ['quit', 'play']
state = ['start', 'stay', 'end']
dice = ['1', '2', '3', '4', '5', '6']
distribution={}
epoch =10000

for i  in range(epoch):
    currentState = 'start'
    totalRewards =0
    while currentState != 'end':

        #newChoice = random.choice(actions)
        randVal = random.random()

        if randVal < 0.999:
            newChoice = 'play'
        else:
            newChoice = 'quit'

        if newChoice =='quit':
            currentState = 'end'
            totalRewards += 10
            continue
        elif currentState == 'start' and newChoice == 'play':
            currentState = 'stay'
            totalRewards += 4
            diceroll = random.choice(dice)
            # or use radom.randint(1,6)
            if diceroll == '1' or diceroll == '2':
                currentState = 'end'
            else:
                currentState = 'stay'
    if totalRewards in distribution:
        distribution[totalRewards] +=1
    else:
        distribution[totalRewards]=1

print(f"Final rewards {totalRewards}")
expectedRewads =0
sortedDist = {k:distribution[k]/epoch for k in sorted(distribution)}
for key, value in sortedDist.items():
    print(f" Reward:{key}, distribution:{sortedDist[key]}")
    expectedRewads += key*sortedDist[key]
print(f"Expected Reward {expectedRewads}")

keys = list(sortedDist.keys())
values = list(sortedDist.values())

# Create a bar plot
plt.bar(keys, values)
plt.title('Keys vs. Values (Bar Chart)')
plt.xlabel('Keys')
plt.ylabel('Values')
plt.grid(axis='y', linestyle='--')  # Optional grid
plt.show()

```
