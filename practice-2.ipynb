```python
import os

def quicksort(userlist):
    if len(userlist)<=1:
        return userlist
    pivot = userlist[len(userlist)//2]
    left = [x  for x in userlist if x < pivot]
    center = [x for x in userlist if x ==pivot]
    right = [x for x in userlist if x >pivot]
    return quicksort(left)+ center + quicksort(right)

testlist = [ 234, 13 , 454, 575 , 23 , 454]

print(quicksort(testlist))
```

```python
import numpy as np

def bubblesort(userlist):
    for i in range(len(userlist)):
        for j in range(0, len(userlist)-i-1):
            if userlist[j] > userlist[j+1]:
                temp = userlist[j]
                userlist[j] = userlist[j+1]
                userlist[j+1]=temp

    return userlist
testlist = [ 234, 13 , 454, 575 , 23 , 454]

print(bubblesort(testlist))

```

```python
import threading

def print_numbers():
    number = [ 1, 2, 3,4, 5, 6, 7]
    for i in number:
        print(i)

def print_char():
    name="adfadfadf"
    for c in name:
        print(c)

# print_numbers()
# print_char()

thread1= threading.Thread(target=print_numbers)
thread2= threading.Thread(target=print_char)

thread1.start()
thread2.start()

thread1.join()
thread2.join()

print("main thread finished")
```

```python
import threading

counter:int =0
counter_lock = threading.Lock()

def incrementCounter(numberList):
    global counter
    for i in numberList:
        with counter_lock:
            counter += i

list1=[1,2,3,4,5,6]
list2=[4,5,6,7,8,9]

thread1 = threading.Thread(target = incrementCounter, args=(list1,))
thread2 = threading.Thread(target = incrementCounter, args=(list2,))

thread1.start()
thread2.start()

thread1.join()
thread2.join()

print("Final value of counter", counter)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

flist=[]

# Generate x and y values
xlist = np.arange(-100, 100, 1)
ylist = np.arange(-100, 100, 1)

# Create a meshgrid for x and y
X, Y = np.meshgrid(xlist, ylist)

# Compute Z as max(X, Y) element-wise
Z = np.maximum(X, Y)

# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis (max(X, Y))')
ax.set_title('3D Plot of Z = max(X, Y)')

# Add color bar for reference
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)
        
# Show the plot
plt.show()

```
